```python
import numpy as np
import pickle
from scalecast.Forecaster import Forecaster
```

```python
sns.set(rc={'figure.figsize':(15,8)})
df = pd.read_csv('COVID-19_Diagnostic_Laboratory_Testing__PCR_Testing__Time_Series (3).csv',parse_dates=['date'])
plt.plot(df_time)
```

```python
f = Forecaster(y=df['new_results_reported'],
                   current_dates=df['date'])
f
```

```python
y = Forecaster(y=df['total_results_reported'],
                   current_dates=df['date'])
y
```

```python
f.plot_pacf(lags=25)
plt.show()
```

```python
y.plot_pacf(lags=25)
plt.show()
```

```python
stat, pval, _, _, _, _ = f.adf_test(full_res=True)
stat
pval
```

```python
#LSTM neural network

f.set_test_length(15)       
f.generate_future_dates(15) 
f.set_estimator('lstm')     
```

```python
y.set_test_length(15)      
y.generate_future_dates(15) 
y.set_estimator('lstm')     
```

```python

f.manual_forecast(call_me='lstm_default')
f.plot_test_set(ci=True)
```

```python
y.manual_forecast(call_me='lstm_default')
y.plot_test_set(ci=True)
```

```python
f.manual_forecast(call_me='lstm_26lags',lags=26)
f.plot_test_set(ci=True)
```

```python
y.manual_forecast(call_me='lstm_26lags',lags=26)
y.plot_test_set(ci=True)
```

```python
f.manual_forecast(call_me='lstm_26lags_5epochs',
                  lags=26,
                  epochs=5,
                  validation_split=.2,
                  shuffle=True)
f.plot_test_set(ci=True)
```

```python
y.manual_forecast(call_me='lstm_26lags_5epochs',
                  lags=26,
                  epochs=5,
                  validation_split=.2,
                  shuffle=True)
y.plot_test_set(ci=True)
```

```python

from tensorflow.keras.callbacks import EarlyStopping
f.manual_forecast(call_me='lstm_26lags_earlystop_3layers',
                  lags=26,
                  epochs=16,
                  validation_split=.2,
                  shuffle=True,
                  callbacks=EarlyStopping(monitor='val_loss',
                                          patience=1),
                  lstm_layer_sizes=(16,16,16),
                  dropout=(0,0,0))
f.plot_test_set(ci=True)
#1hr 
```

```python
f.manual_forecast(call_me='lstm_best',
                  lags=35,
                  batch_size=30,
                  epochs=15,
                  validation_split=.2,
                  shuffle=True,
                  activation='tanh',
                  optimizer='Adam',
                  learning_rate=0.001,
                  lstm_layer_sizes=(76,)*2,
                  dropout=(0,)*2,
                  plot_loss=True)
f.plot_test_set(order_by='LevelTestSetMAPE',models='top_2',ci=True)
```

```python
f.set_estimator('mlr') 
f.add_ar_terms(26) 
f.add_seasonal_regressors('month','quarter',dummy=True) 
f.add_seasonal_regressors('year') 
f.add_time_trend() 
f.diff() 
```

```python
f.manual_forecast()
f.plot_test_set(order_by='LevelTestSetMAPE',models='top_2')
```

```python
f.plot_test_set(models='mlr',ci=True)
```

```python
f.plot(models=['mlr','lstm_best'],
       order_by='LevelTestSetMAPE',
       level=True)
```

```python
f.export('model_summaries',determine_best_by='LevelTestSetMAPE')[
    ['ModelNickname',
     'LevelTestSetMAPE',
     'LevelTestSetRMSE',
     'LevelTestSetR2',
     'best_model']
]
```
